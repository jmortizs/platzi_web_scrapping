```python
def get_discography(artist_id, token, return_name=False, page_limit=50, country=None):
    url = f'{base_url}/artists/{artist_id}/albums'
    header = {'Authorization': f'Bearer {token}'}
    params = {'limit': page_limit, 'offset': 0, 'country': country}
    
    result = []
    r = requests.get(url, params=params, headers=header)
    
    if r.status_code != 200:
        print('Request error')
        return None
    
    if return_name:
        result += [(item['id'], item['name']) for item in r.json()['items']]
    else:
        result += [item['id'] for item in r.json()['items']]
        
    while r.json()['next']:
        r = requests.get(r.json()['next'], headers=header)
        if return_name:
            result += [(item['id'], item['name']) for item in r.json()['items']]
        else:
            result += [item['id'] for item in r.json()['items']]
            
    return result
    
def get_tracks(album_id, token, return_name=False, page_limit=50, market=None):
    url = f'{base_url}/albums/{album_id}/tracks'
    header = {'Authorization': f'Bearer {token}'}
    params = {'limit': page_limit, 'offset': 0, 'market': market}
    
    result = []
    r = requests.get(url, params=params, headers=header)
    
    if r.status_code != 200:
        print('Request error')
        return None
    
    if return_name:
        result += [(item['id'], item['name']) for item in r.json()['items']]
    else:
        result += [item['id'] for item in r.json()['items']]
        
    while r.json()['next']:
        r = requests.get(r.json()['next'], headers=header)
        if return_name:
            result += [(item['id'], item['name']) for item in r.json()['items']]
        else:
            result += [item['id'] for item in r.json()['items']]
            
    return result        
    
```

```python
for album in get_discography(artist_id, access_token, return_name=True, country='CO'):
    print(f'{album[1]} ({album[0]})')
    for track in get_tracks(album[0], access_token, return_name=True, market='CO'):
        print('  ', track[1])
```
